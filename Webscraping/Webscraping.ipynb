```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
webpage = requests.get('https://quotes.toscrape.com/').text
```

```python
soup =BeautifulSoup(webpage,'lxml')
```

```python
qt = soup.find_all('div',class_='quote')
```

```python
len(qt)
```

```python
quote=[]
author=[]
# tag = []
for i in qt:
    quote.append(i.find_all('span')[0].text.strip())
    author.append(i.find_all('span')[1].text.strip())
    # tag.append(i.find('meta',class_='keywords').text)
```

```python
tags = []

for i in soup.find_all('div', class_='tags'):
    clean = i.text.replace("Tags:", "").strip()
    one_line = ", ".join(t.strip() for t in clean.split("\n") if t.strip())
    tags.append(one_line)
```

```python
dict = {'quote':quote, 'author':author,'tags':tags}
```

```python
df = pd.DataFrame(dict)
```

```python
final = pd.DataFrame()

for j in range(1,11):
    webpage = requests.get('https://quotes.toscrape.com/').text
    soup =BeautifulSoup(webpage,'lxml')
    qt = soup.find_all('div',class_='quote')
    quote=[]
    author=[]
    for i in qt:
        quote.append(i.find_all('span')[0].text.strip())
        author.append(i.find_all('span')[1].text.strip())
    tags = []
    for i in soup.find_all('div', class_='tags'):
        clean = i.text.replace("Tags:", "").strip()
        one_line = ", ".join(t.strip() for t in clean.split("\n") if t.strip())
        tags.append(one_line)

    dict = {'quote':quote, 'author':author,'tags':tags}

    df = pd.DataFrame(dict)
    
    # final =final.append(df)

    final = pd.concat([final, df], ignore_index=True)
```

```python
final.to_csv("quotes_data.csv", index=False)
```
